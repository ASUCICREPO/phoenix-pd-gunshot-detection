```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("GunShotURL.xlsx")
```

```python
df.rename(columns = {'Type of Sound (Gunshot, Unclear, etc.)':'Class'}, inplace = True)
```

```python
for index, row in df.iterrows():
    if row['Class'] == "Possible Gunshot" or row['Class'] == "Gunshot" :
        df.at[index,'Class'] = "Gunshot"
    else :
        df.at[index,'Class'] = "Not Gunshot"
```

```python
print(df['Class'].unique())
```

```python
df['Class'].value_counts()
```

```python
one_hot_encoded_data = pd.get_dummies(df, columns = ['Class'])
print(one_hot_encoded_data)
```

```python
one_hot_encoded_data.to_excel("Gunshotencoded.xlsx")  
```

```python
import wget

for index, row in one_hot_encoded_data.iterrows():
    URL = row['File URL']
    fileName = str(index) + '.wav'
    response = wget.download(URL, fileName)
```
